```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.1. Скачать данные по ссылке https://www.kaggle.com/datasets/ionaskel/laptop-prices

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
df.sample(n=5)
```

## 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(20, 20))

sns.histplot(df['price'], bins=15)

plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Кол-во');
```

## 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(10, 8))

sns.histplot(df['sqft_living'], bins=15)

plt.title('Распределение жилой квадратуры')
plt.xlabel('Квадратура')
plt.ylabel('Кол-во');
```

## 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10, 8))

sns.histplot(df['yr_built'], bins=15)

plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Кол-во');

```

```python
data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(data, autopct='%1.1f%%')

plt.title('Распределение домов от наличия вида на набережную')
plt.legend(data.index);
```

# Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Из данного графика можно сделать вывод, что на стоимость влияет площадь.

```python
sns.jointplot(x=df['price'], y=df['condition'], kind='reg');
```

Из данного графика видно, что состояние практически не влияет на стоимость. Большинство квартир в среднем состоянии.

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
```

Из графика видно, что год постройки влияет на стоимость: цена за дома, построенные в 2000х годах цена выше, чем в 1950 и ранее.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg');
```

Из графика видно, что количество комнат не влияет на цену. Квартиры с большим количеством комнат находятся находятся в нижнем ценовом сегменте.
